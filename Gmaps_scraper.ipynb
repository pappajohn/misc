```python
## find and import the CSV

import os
os.chdir('/Users/Arjun/Desktop/python_scripts')

import pandas as pd
df = pd.read_csv("restaurants20211016.csv", sep=",")
#read in df

df = df.drop(columns=['google_place_id','place_id','raw_data','location','review_count','rating'])
#drop extra columns

df_sorted = df.sort_values(by=['weighted_rating'], ascending=False)
#sort by descending rank

df_sorted['Cuisine'] = 'x'
#add fourth column for cuisine

df_list = df_sorted.values.tolist()
#make it a list for use later

## create script to scrape and find the cuisine type, or return an error if not there

import requests
from bs4 import BeautifulSoup

from random import randint
from time import sleep
```

```python
def get_category(url):
    url = str(url)

    try:
        print('make page')
        page = requests.get(url)
        #print(page.text)
        print()

        print("get soup")
        soup = BeautifulSoup(page.content, "html.parser")
        #print(soup.prettify)
        print()

        print("get title")
        title = soup.find("meta", property="og:description")
        print(str(title))
        print()

        print("split title")
        a = str(title).split("· ",1)[1]
        # remove everything before "· "

        b = a.split("\"",1)[0]
        # remove everything after the first quote. \ needed because three consecutive double quotes break split

    except:
        b = str("script_error")

    print(b)
    print()
    return b
```

```python
def append_categories(restolist, num_rows):

    counter = 0
    cuisine_list = []

    for resto in restolist:
        print('start ' + resto[0])
        print('counter @ ' + str(counter))
        print(cuisine_list)

        if resto[3] == 'x':

            timelag = randint(5, 15)
            print('wait ' + str(timelag) + ' secs')
            sleep(timelag)
            #wait a bit for rate limiting

            url = resto[1]
            cat = get_category(url)
            resto[3] = (cat)
            print(resto)
            counter = counter + 1
            cuisine_list.append(cat)

        if cuisine_list.count('script_error') >= 4:
            break

        if counter == num_rows:
            break
```
